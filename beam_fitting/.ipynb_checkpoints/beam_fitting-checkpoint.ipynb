```python
from sotodlib import core
import numpy as np
import matplotlib.pyplot as plt
import so3g
from scipy.optimize import curve_fit
from sotodlib.io.load import load_file
from glob import glob
import time 
%matplotlib notebook
```

# Calibration simulation

```python
context = core.Context('pipe-sim-metadata/pipe-s0001.yaml')
files = sorted(glob('/mnt/so1/users/aamirmali/bcp_sims/genesis_lt2_course_atm_wn_tau/CES-Atacama-LAT-Jupiter-0-0/MFF2*.g3'))
print('%d files found.'%len(files))
tod = load_file(files)
meta = context.get_meta({}) # you might need a timestamp or something in there!
tod.merge(meta)
```

```python
ctime = tod['timestamps']
data = tod['signal']

az = tod['boresight']['az']
el = tod['boresight']['el']

csl = so3g.proj.CelestialSightLine.az_el(ctime, az, el, weather='typical')
q_bs = csl.Q
assembly = so3g.proj.Assembly.attach(csl, tod.focal_plane.quat)
ra, dec = np.transpose(csl.coords()[:, :2])
```

```python
# Finding the object
import ephem
from datetime import datetime
from so3g.proj import coords, quat
from astropy.table import Table
timestamp = np.mean(ctime)
dt = datetime.fromtimestamp(timestamp)
dt_str = dt.strftime('%Y/%m/%d %H:%M:%S')
print(dt_str)
obj = ephem.Jupiter()
obj.compute(dt_str)
ra_obj = obj.ra
dec_obj = obj.dec
```

```python
# Peak-finding for initial guess
sig_peak = []
dets = []
xi_offset = []
eta_offset = []
for det in range(data.shape[0]):
    data_det = data[det, :]
    sig_peak.append(np.max(data_det))
    
peak_cut = 1 #HARD CODED!!!
for det in range(data.shape[0]):
    #print(det, end=' ')
    data_det = data[det, :]
    percentile = 99.99
    #if np.max(data_det) < np.mean(sig_peak):
    if np.max(data_det) < peak_cut:
        continue
    thresh = np.nanpercentile(data_det, percentile)
    idx_t = np.where(data_det > thresh)
    ra_peaks = np.mod(ra[idx_t], 2*np.pi)
    dec_peaks = dec[idx_t]
    #print(np.std(ra_peaks), np.std(dec_peaks))
    ra_bs = np.mean(ra_peaks)
    dec_bs = np.mean(dec_peaks)
    
    q_obj = quat.rotation_lonlat(np.mean(ra_obj), 
                                 np.mean(dec_obj))
    q_bs_peak = coords.quat.quat(*np.average(np.array(q_bs)[idx_t], axis=0))
    xi_t, eta_t, _ = quat.decompose_xieta(~q_bs_peak * q_obj)
    xi_offset.append(xi_t)
    eta_offset.append(eta_t)
    dets.append(det)
det_table = Table([dets, xi_offset, eta_offset], names=['det_uid', 'xi', 'eta'],
                  dtype=['i4', 'f4', 'f4'])
```

```python
q_obj = quat.rotation_lonlat(np.mean(ra_obj), np.mean(dec_obj))
q_bs = csl.Q
xi_t, eta_t, _ = quat.decompose_xieta(~q_bs*q_obj)
```

```python
fig3, ax3 = plt.subplots(ncols=1, figsize=(4, 4))
ax3.scatter(np.degrees(xi_offset), np.degrees(eta_offset), s=5)
ax3.scatter(np.degrees(xi_t), np.degrees(eta_t), s=1, alpha=0.01)
ax3.set_aspect('equal')
ax3.set_xlabel(r'$\xi$ (deg)')
ax3.set_ylabel(r'$\eta$ (deg)')
ax3.set_title('Focal Plane Pointing')
fig3.tight_layout()
```
